## Template: Equilibrium Stoichiometry + Delta G Questions

Use this flow for questions like:
1. Write stoichiometric coefficients (negative for reactants, positive for products).
2. Convert Delta G degree to K with $K = e^{-\Delta G^\circ/(RT)}$ (if needed).
3. Use extent of reaction $\xi$ with $n_i = n_{i,0} + \nu_i\xi$.
4. Solve $\xi$ from a species where $n_{i,0}$ is known/assumed.
5. Compute unknown equilibrium amount(s).
6. Optional check: compute $Q$ from equilibrium amounts and compare with $K$.

For 1 L, moles and molarity are numerically the same.

```python
from math import exp

R = 8.314  # J/(mol*K)

def delta_g_to_K(delta_g_kj_per_mol, T_K=298.15):
    """Convert Delta G degree (kJ/mol) to equilibrium constant K."""
    delta_g_j = delta_g_kj_per_mol * 1000.0
    return exp(-delta_g_j / (R * T_K))

def extent_from_species(n_eq, nu_i, n0_i=0.0):
    """Solve extent xi from n_i(eq) = n_i0 + nu_i*xi."""
    return (n_eq - n0_i) / nu_i

def n_eq_from_extent(xi, nu_i, n0_i=0.0):
    """Compute equilibrium amount from extent xi."""
    return n0_i + nu_i * xi

# -----------------------------
# Reusable template input block
# -----------------------------
# Example reaction: 2A + B <=> 2C + D
nu = {"A": -2, "B": -1, "C": 2, "D": 1}

T = 298.0
delta_g0_kj = -10.0

# Known equilibrium amounts in 1 L (mol == M numerically)
n_eq_known = {"A": 0.10, "B": 0.15, "C": 0.65}

# Assumption for this style of question: product used for xi started at 0
n0_assumed = {"C": 0.0, "D": 0.0}

# 1) Thermodynamic constant from Delta G degree (optional but useful for checks)
K = delta_g_to_K(delta_g0_kj, T)

# 2) Solve extent from known product C½
xi = extent_from_species(n_eq_known["C"], nu["C"], n0_assumed["C"])

# 3) Solve unknown D at equilibrium
n_D_eq = n_eq_from_extent(xi, nu["D"], n0_assumed["D"])

# 4) Optional consistency check with Q
Q = (n_eq_known["C"]**2 * n_D_eq) / (n_eq_known["A"]**2 * n_eq_known["B"])

print(f"K from Delta G degree: {K:.3f}")
print(f"Extent xi: {xi:.4f} mol")
print(f"D at equilibrium: {n_D_eq:.4f} mol")
print(f"Q from equilibrium amounts: {Q:.3f}")

# -----------------------------
# Quick-use mini-template
# -----------------------------
# Replace these values for new questions:
## nu = {"A": ..., "B": ..., ...}
## n_eq_known = {"species": value, ...}
## n0_assumed = {"species": value, ...}
## xi = extent_from_species(n_eq_known["known_species"], nu["known_species"], n0_assumed.get("known_species", 0.0))
## n_unknown = n_eq_from_extent(xi, nu["unknown_species"], n0_assumed.get("unknown_species", 0.0))
```
