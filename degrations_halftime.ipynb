```python
import equilibrium as eq
from compound import Compound
from equations import ChemicalEquation

import math

def get_k_from_delta_g(delta_g_kj, temp_k):
    """Calculates K using delta_g = -RT ln K"""
    # R in J/(mol·K), so convert kJ to J
    r = 8.314
    delta_g_j = delta_g_kj * 1000
    # Rearrange to: K = e^(-delta_g / RT)
    exponent = -delta_g_j / (r * temp_k)
    return math.exp(exponent)

def solve_for_d(k, a_conc, b_conc, c_conc):
    """
    K = ([C]^2 * [D]) / ([A]^2 * [B])
    Solving for [D]:
    [D] = (K * [A]^2 * [B]) / [C]^2
    """
    numerator = k * (a_conc**2) * b_conc
    denominator = c_conc**2
    return numerator / denominator
```

```python
# --- Given Values from Image ---
delta_g_standard = -10  # kJ/mol
temperature = 298       # K
moles_a = 0.1           # M (since 1 L)
moles_b = 0.15          # M (since 1 L)
moles_c = 0.65          # M (since 1 L)

# --- Calculation ---
# 1. Find the equilibrium constant
k_value = get_k_from_delta_g(delta_g_standard, temperature)

# 2. Solve for the missing concentration of D
concentration_d = solve_for_d(k_value, moles_a, moles_b, moles_c)

print(f"Equilibrium Constant (K): {k_value:.2f}")
print(f"Amount of D at equilibrium: {concentration_d:.2f} mol")
```
